```python
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
import numpy as np
from IPython.display import HTML # Needed to display the animation in a Notebook

# 1. Initialization
fig, ax = plt.subplots()
x_data, y_data = [], []
line, = ax.plot([], [], 'r-') # Comma is important! It unpacks the single line object

# Set axis limits
ax.set_xlim(0, 2 * np.pi)
ax.set_ylim(-1.5, 1.5)

# 2. Update Function (The "Animator")
def update(frame):
    """Called once per frame to update the data."""
    t = frame * 0.1 # Increase time step
    
    # Update the data lists
    x_data.append(t)
    y_data.append(np.sin(t))
    
    # Check if we should reset the plot
    if t > 2 * np.pi:
        x_data.clear()
        y_data.clear()

    # Update the Line object's data
    line.set_data(x_data, y_data)
    
    return line, # Must return the artists that were modified

# 3. Create and Run the Animation
# interval is in milliseconds
anim = FuncAnimation(fig, update, frames=np.arange(0, 100), interval=50, blit=True) 

# Display the animation in the Jupyter Notebook
# NOTE: This may require ffmpeg or other encoders to be installed for complex animations.
# The `to_jshtml()` method is often more reliable in a notebook environment.
plt.close(fig) # Prevent the static figure from showing twice
HTML(anim.to_jshtml())
```
