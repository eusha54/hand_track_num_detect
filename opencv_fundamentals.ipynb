```python
# a basic video capture example (esc to exit)

import cv2

cap = cv2.VideoCapture(0)
    # 0 usually default camera
    # a video capture object

if not cap.isOpened():
    # checks if caputure object initialized successfully
    raise Exception('Camera not found')

cap_read = cap.read()
    # fetches the next frame in the stream
    # returns tuple(ret, frame)
        # ret : True if frame successfully read else False
        # frame: a 3d np array (pixel, pixel, BGR color format)
            # like ((480, 640, 3))


# capturing video (esc to exit)
while cap.isOpened():
    ret, frame = cap.read()

    if not ret:
        break

    frame = cv2.flip(frame, 1)
        # fliping image for mirror like experience

    cv2.imshow("frame", frame)
        # takes window_name(str) and img frame(3d np arr)
        # then renders the image on the window
    
    if cv2.waitKey(delay=1) == 27:
        # controls the window refresh rate
        # delay in milisecond
        # returns ascii code of keypress while window is on
            # 27 = esc

        break
            # closing the loop

cap.release()
    # releasing video handler

cv2.destroyWindow("frame")
    # destroys the window named "frame"
```

```python
# a basic video writer example
import cv2

cap = cv2.VideoCapture(0)
fourcc = cv2.VideoWriter_fourcc(*'mp4v')
    # four character code (fourcc)
    # 'mp4v' is codec name
    # asterick is is unpacking argument
    # same as cv2.VideoWriter_fourcc('m','p','4','v')
    # *'mp4v' expands into 'm','p','4','v'
    # cv2.VideoWriter_fourcc converts those 4 chars into a 32-bit integer
    # That integer is required by cv2.VideoWriter to pick the correct video codec
    
out = cv2.VideoWriter("out.mp4", fourcc, 30, (640, 480))

while cap.isOpened():
    ret, frame = cap.read()
    
    if not ret:
        break
    
    frame = cv2.flip(frame, 1)

    out.write(frame)
    cv2.imshow("Recording", frame)

    if cv2.waitKey(1) == 27:
        break

cap.release()
out.release()
cv2.destroyAllWindows()
```
