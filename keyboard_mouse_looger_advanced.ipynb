```python
from pynput import keyboard

def on_press(key):
    try:
        print(f'{key.char} pressed')
    except AttributeError:
        print(f'{key} pressed')

def on_release(key):
    if key == keyboard.Key.esc:
        return False

with keyboard.Listener(
    on_press = on_press,
    on_release = on_release
    ) as listener:
    listener.join()
```

```python
from pynput import mouse 

def on_move(x, y):
    print(f'mouse moved to ({x},{y})')

def on_click(x, y, button, pressed):
    action = 'pressed' if pressed else 'released'
    print(f'{button} : {action} at ({x},{y})')

    if button == mouse.Button.middle and not pressed:
        return False
        # stop listener after one click on middle mouse button

with mouse.Listener(
    on_move=on_move,
    on_click=on_click
    ) as listener:
    listener.join()
```
